### Understanding `tf.newaxis`

`tf.newaxis` is used to increase the dimension of an existing tensor (or NumPy array) by one unit. It's essentially a placeholder that creates a new axis (dimension) at the position where it's inserted. This is particularly useful when you need to broadcast tensors for element-wise operations or when a function expects a tensor of a specific rank (number of dimensions).

```python
import tensorflow as tf

# Let's start with a 1-D tensor
vector = tf.constant([1, 2, 3, 4])
print("Original Vector:", vector.numpy())
print("Shape of Original Vector:", vector.shape)
print("Number of dimensions (rank):", vector.ndim)
```

### Example: Adding a new axis

```python
# 1. Adding a new axis at the beginning (making it a row vector, shape (1, 4))
reshaped_vector_1 = vector[tf.newaxis, :]
print("\nReshaped Vector (tf.newaxis, :):")
print(reshaped_vector_1.numpy())
print("Shape:", reshaped_vector_1.shape)
print("Number of dimensions (rank):", reshaped_vector_1.ndim)
```

```python
# 2. Adding a new axis at the end (making it a column vector, shape (4, 1))
reshaped_vector_2 = vector[:, tf.newaxis]
print("\nReshaped Vector (:, tf.newaxis):")
print(reshaped_vector_2.numpy())
print("Shape:", reshaped_vector_2.shape)
print("Number of dimensions (rank):", reshaped_vector_2.ndim)
```

```python
# You can also use tf.expand_dims for the same purpose, which is often more explicit.
# Example: Adding a dimension at index 0 (beginning)
expanded_vector_1 = tf.expand_dims(vector, axis=0)
print("\nExpanded Vector (tf.expand_dims, axis=0):")
print(expanded_vector_1.numpy())
print("Shape:", expanded_vector_1.shape)

# Example: Adding a dimension at index 1 (end)
expanded_vector_2 = tf.expand_dims(vector, axis=1)
print("\nExpanded Vector (tf.expand_dims, axis=1):")
print(expanded_vector_2.numpy())
print("Shape:", expanded_vector_2.shape)
```
