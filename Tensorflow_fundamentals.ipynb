create randor vector

```python
random_1=tf.random.Generator.from_seed(42)
random_1=random_1.normal(shape=(3,2))
random_1
```

```python
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))
random_2
```

To randomize the inpur data stream in a NN network, Vector/Tensor should be shuffled.
Lets shuffle a Vector/Tensor

```python
tf.random.shuffle(random_1)
```

Notice that `random_2` is identical to `random_1` because the generator was re-initialized with the same seed. If you want a *different* set of random numbers with a different seed, you would use a different seed value.
